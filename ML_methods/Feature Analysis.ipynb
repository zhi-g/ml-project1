## Load the training data into feature matrix, class labels, and event ids:

```python
from proj1_helpers import *
DATA_TRAIN_PATH = '../Data/train.csv' # TODO: download train data and supply path here 
y, tX, ids = load_csv_data(DATA_TRAIN_PATH)
```

```python
for i in range(30):
    print("Feature " + str(i))
    f = tX[:,i] != -999
    plt.scatter(tX[f, i], y[f])
    plt.savefig("../Figures/Feature" + str(i) + ".png")
    plt.show()
```

```python
for i in range(30):
    print("Feature " + str(i))
    f = tX[:,i] != -999
    b = y[:] == 1
    s = y[:] != 1
    plt.figure(figsize=(20,5))
    plt.subplot(121)
    plt.hist(tX[np.logical_and(f, b), i], 50, facecolor = "blue")
    plt.subplot(122)
    plt.hist(tX[np.logical_and(f, s), i], 50, facecolor = "red")
    plt.savefig("../Figures/Histogram/Feature" + str(i) + ".png")
    plt.show()
```

```python
for i in range(30):
    f = tX[:,i] == -999
    print("Feature " + str(i) + " -999: " + str(np.shape(tX[f,i])[0] / np.shape(tX)[0] * 100) + "%")
```
